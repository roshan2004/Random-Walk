The Random 1D program simulates a random walk in one dimension for steps of length 1. The default number of steps is N = 100 and the probability p of going to the right for a single step is 0.5, and to the left is 1-0.5 = 0.5.

```python
import random
import numpy
import matplotlib.pyplot as plt
                     # no of particles = 1
ns = 100                        # no of steps
positions = [0]     # all particles start at x=0
HEAD = 0;  TAIL = 1         # constants

for step in range(ns):
        coin = random.randint(0,1)  # flip coin
        if coin == HEAD:
            positions.append(positions[-1]+1)  # one unit length to the right
        elif coin == TAIL:
            positions.append(positions[-1]-1)   # one unit length to the left
```

```python
plt.plot(positions, linestyle ='', marker = 'o')
```

```python
import numpy
import random
HEAD = 1
TAIL = 2



a = numpy.zeros((2, 50))
b = a.transpose()
ns = 50                        # no of steps

for step in range(ns):
    for _ in range(ns):
        coin_x = random.randint(1,2)

        # Move on x axis
        if coin_x == HEAD:
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1   

        # Move on y axis
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1 
        elif coin_y == TAIL:
           b[step,1]-=1  
```

```python
plt.scatter(b[:,0], b[:,1])
```

```python
import numpy
import random
HEAD = 1
TAIL = 2



a = numpy.zeros((3, 100))
b = a.transpose()
ns = 100                       # no of steps

for step in range(ns):
    for _ in range(ns):
        coin_x = random.randint(1,2)

        # Move on x axis
        if coin_x == HEAD:
           b[step,0]+=1  
        elif coin_x == TAIL:
           b[step,0]-=1   

        # Move on y axis
        coin_y = random.randint(1,2)
        if coin_y == HEAD:
           b[step,1]+=1 
        elif coin_y == TAIL:
           b[step,1]-=1  
        
        #move on z axis
        coin_z = random.randint(1,2)
        if coin_z == HEAD:
           b[step,2]+=1 
        elif coin_z == TAIL:
           b[step,2]-=1 
```

```python
%matplotlib notebook
fig_3d = plt.figure(figsize = (10,8))
fig_3d.add_subplot(1,1,1,projection="3d")
ax = plt.gca()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.plot(b[:,0],b[:,1],b[:,2], linestyle = '--')
plt.show()
```
